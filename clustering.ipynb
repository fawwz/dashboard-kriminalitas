```python
df = pd.read_excel('2018.xlsx')
provinsi = df['Provinsi']
df = df.drop(columns=[
    'Provinsi','Jumlah Kejahatan (Crime Total) yang dilaporkan Menurut Polda',
    'crime rate',
    'kejahatan terhadap ketertiban umum'
    ])
```

```python
df.isnull().sum()n 
```

Modeling (K-Means)

```python
kmeans = KMeans(n_clusters=3,random_state=144)
kmeans.fit(df)
```

```python
labels = kmeans.labels_
```

```python
X = df[['asusila', 'kejahatan terhadap nyawa', 'kejahatan terhadap kemerdekaan', 'kejahatan terhdap hak milik', 'kejahatan terhadap hak milik dengan kekerasan', 'kejahatan terkait narkotika', 'kejahatan penipuan, penggelapan, dan korupsi', 'kejahatan terhadap fisik']]
```

```python
labels
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(X,labels)
print("Silhouette Score:", silhouette_avg)
```

```python
df['cluster'] = labels
df.head()
```

Cluster profiling

```python
df_0 = df.loc[df['cluster']==0]
df_1 = df.loc[df['cluster']==1]
df_2 = df.loc[df['cluster']==2]
```

```python
df_2.describe().mean().mean()
```

0 = Rendah

1 = Tinggi

2 = Sedang

```python
tingkat_kriminalitas_bin=[]
for i in df['cluster']:
    if i == 0 :
        tingkat_kriminalitas_bin.append('Rendah')
    elif i == 1 :
        tingkat_kriminalitas_bin.append('Tinggi')
    else:
        tingkat_kriminalitas_bin.append('Sedang')
```

```python
df['cluster'] = tingkat_kriminalitas_bin
df = df.rename(columns={'cluster': 'Tingkat Kriminalitas'})

df.insert(0, 'Provinsi',provinsi)
df
```

```python
df.to_excel('2018_FINAL.xlsx',index=False)
```
